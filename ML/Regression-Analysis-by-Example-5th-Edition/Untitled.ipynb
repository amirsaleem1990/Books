```python
url = "http://www1.aucegypt.edu/faculty/hadi/RABE5/#Download"
import requests
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(requests.get(url).text, "lxml")
```

```python
for i in soup.find_all("center")[2]:
    try:
        print(i.find_all("table", {"cellspacing" : "0"})[0].find("tr"))
    except:
        pass
```
